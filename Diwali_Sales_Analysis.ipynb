```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.dropna(inplace=True) #Dropping null Values
```

```python
ax = sns.countplot(x = 'Gender', color='lightblue', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
each_sale = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
colors = ['orange' if gender == 'M' else 'lightblue' for gender in each_sale['Gender']]
sns.barplot(x= 'Gender', y='Amount', hue='Gender', data= each_sale, palette= colors,legend=False)
plt.show()
```

```python
ax = sns.countplot(x = 'Age Group', hue='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x= 'Age Group', y='Amount',hue='Age Group', palette='Set2', data= sales_age,legend=False)
plt.show()
```

```python
#By Orders
sale_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'State', y='Orders', hue='State', palette='Set1',data= sale_states,legend=False)
plt.show()
# By Amount
sale_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'State', y='Amount', hue='State',palette='Set2', data= sale_states,legend=False)
plt.show()
```

From Above Graph We can conclude that most of the sales are from UP, Maharashtra and Karnatake respectively.

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc={'figure.figsize':(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sale_states = df.groupby(['Marital_Status' ,'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x= 'Marital_Status', y='Amount', hue='Gender', data= sale_states)
plt.show()
```

From above graph we can conclude that most of the buyers are married women and they have high purchasing power.

```python
sns.set(rc={'figure.figsize':(20,7)})
ax = sns.countplot(x = 'Occupation', hue='Occupation',data = df, palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
Sales_Occupation= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(x='Occupation', y='Amount', hue='Occupation', palette='Set2', data=Sales_Occupation)
plt.show()
```

Conclusion : Most of the buyers are from IT, Healthcare and Aviation sector.

```python
sns.set(rc={'figure.figsize':(25,7)})
ax = sns.countplot(x = 'Product_Category', hue='Product_Category', palette='Set1', data = df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
Sales_Occupation= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(x='Product_Category', y='Amount', hue='Product_Category', palette='Set2', data=Sales_Occupation)
plt.show()
```

Insight : Most of the sold products are from Food, Clothing and Electronics category.

```python
Sales_Occupation= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(x='Product_ID', y='Orders', hue='Product_ID', palette='Set2', data=Sales_Occupation)
plt.show()
```

Summery : Married Women age group 26-35 years from UP, Maharastra and Karnatake Working in IT, Healthcare and Aviation are more likely to buy products from food clothing and Electronics Category.
